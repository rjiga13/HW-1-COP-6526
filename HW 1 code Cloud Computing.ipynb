```python
sc = pyspark.SparkContext()
```

```python
file = sc.textFile("input.txt")
```

```python
def pairs (words: list, key) -> list:
    summary = list()
    copies = words.copy()
    copies.remove(key)
    for word in copies:
        summary.append((key, word))
    return summary
```

```python
def findPairs (words: list) -> list:
    Dict = {}
    for word in words:
        if word not in Dict.keys():
            Dict[word] = pairs(words, word)
    return list(Dict.values())
```

```python
## remove all values except for a-z and space, assign to regex variable
```

```python
regex = re.compile("[^a-zA-Z\s]")
```

```python
finalCounts = file\
    .map(lambda x: x.lower())\
    .map(lambda x: regex.sub("", x))\
    .map(lambda x: x.split())\
    .flatMap(lambda x: findPairs(x)).flatMap(lambda x: x)\
    .map(lambda x: (x, 1))\
    .reduceByKey(lambda x, y: x + y)\
    .sortByKey()
```

```python
count = 1
```

```python
## Count value is to the right of the pair
```

```python
for pair in finalCounts.collect():
    print(pair[0], pair[1])
    count += 1
```
